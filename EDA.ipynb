```python
import ast

def extract_genres(genres):
    try:
        genres = ast.literal_eval(genres)
        return [g['name'] for g in genres]
    except:
        return []

df['genre_list'] = df['genres'].apply(extract_genres)
```

```python
def watched_with(genres):
    if 'Animation' in genres or 'Family' in genres:
        return 'Family'
    elif 'Romance' in genres:
        return 'Partner'
    elif 'Comedy' in genres or 'Adventure' in genres:
        return 'Friends'
    elif 'Horror' in genres or 'Thriller' in genres or 'Crime' in genres:
        return np.random.choice(['Friends', 'Alone'])
    else:
        return 'Alone'

df['Watched_With_Whom'] = df['genre_list'].apply(watched_with)
```

```python
sns.countplot(x='Watched_With_Whom', data=df)
plt.title("Watched With Whom Distribution")
plt.show()
```

```python
df.groupby('Watched_With_Whom')['vote_average'].mean()
```

```python
sns.barplot(
    x='Watched_With_Whom',
    y='vote_average',
    data=df
)
plt.title("Average Rating by Watching Context")
plt.show()
```

```python
sns.boxplot(
    x='Watched_With_Whom',
    y='popularity',
    data=df
)
plt.title("Popularity by Watching Context")
plt.show()
```

```python
df.groupby('Watched_With_Whom')['budget'].mean()
```

```python
sns.barplot(
    x='Watched_With_Whom',
    y='budget',
    data=df
)
plt.title("Average Budget by Watching Context")
plt.show()
```

```python
numeric_df = df[['budget', 'revenue', 'popularity', 'vote_average']]
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
df['is_family'] = df['Watched_With_Whom'].apply(lambda x: 1 if x == 'Family' else 0)
df['is_friends'] = df['Watched_With_Whom'].apply(lambda x: 1 if x == 'Friends' else 0)
df['is_partner'] = df['Watched_With_Whom'].apply(lambda x: 1 if x == 'Partner' else 0)
df['is_alone'] = df['Watched_With_Whom'].apply(lambda x: 1 if x == 'Alone' else 0)
```

```python
df.to_csv("tmdb_movies_with_watching_context.csv", index=False)
```
